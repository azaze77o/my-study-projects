## Предобработка данных

```python
df.isna().mean().sort_values(ascending=False)
```

Самую высокую долю по пропускам занимают Аппартаменты - 88%.  
Можно предположить что пропуски свяpаны с тем, что это специфичный вид недвижимости, встречается в объявлениях реже типичной квартиры, поэтому пропуск означают то, что объект не относится к апартаментам. Простави по ним нули.  

Также имеем значительную долю пропусков в данных по расстояниям: до парка и до водоема *(66% и 62 % соответственно)*,  до центра и до аэропорта *(23%)*. Эти данные получены на системном уровне, и их заполнение является не обоснованным, т.к. в понимании рассстояния, необходимы точные величины. О такого рода проблемах сообщим разработке. 

Наименьшую долю по пропускам имеют:  
**до 1%**: этажность и название населеннго пункта. В исслеоваднии не играют большой роли, заполнять не будем.  
**8 - 9%**: площадь кухни и площадь жилого помещения

Разделим пропуски на категории, те которые будем заполнять и нет.
***
Заполняем:
- Апартаменты *(is_apartment)*
- Число балконов *(balcony)*
- Высота потолков *(ceiling_height)*


Не заполняем:
- Расстояние до парка *(parks_nearest)*
- Расстояние до водоема *(ponds_nearest)*
- Расстояние до аэоропорта *(airports_nearest)*
- Расстояние до центра *(cityCenters_nearest)*
- Число дней активности объявления *(days_exposition)*
- Этажность *(floors_total)*
- Название населенного пунтка *(locality_name)*
- Число парков *(parks_around3000)*
- Число водоемов *(ponds_around3000)*
- Площадь кухни *(kitchen_area)*
- Жилая площадь *(living_area)*


```python
# Определеяем среднее значение для высоты потолка с ограничем выборки от 2 до 5 метров.
mean_ceiling_height = df[(df['ceiling_height'] > 2) | (df['ceiling_height'] < 5)] ['ceiling_height'].mean()
```

```python
# Заполняем пропуски по высоте потолка
df['ceiling_height'] = df['ceiling_height'].fillna(mean_ceiling_height)
```

```python
# Заполним пропуски по типу "Апартаменты" и переведем тип данных в логический тип
df['is_apartment'] = df['is_apartment'].fillna(0).astype(bool)
```

```python
# Заполним пропуски  по числу баклонов и переведем тип данных в целочисленный
df['balcony'] = df['balcony'].fillna(0).astype(int)
```

```python
# Переведем дату публикации в нужный формат: datetime.
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

### Вывод
***
1. Мы выполнили частичное **заполнение данных с пропусками**, там где это логично, большую из них оставили без изменений.  
   Это либо данные полученные на системном уровне, восставновление которых проблематично  
   *(расстояния, определенные по геолокации).*  
   Либо данные, отсутствие которых не случайно.
   
   
2. Произведено **приведение данных к нужным типам**:  
   - Дата публикации приведена к типу *datetime* для возможности выделить день недели, месяц, год.
   - Апартаменты приведены к булевому значению, так после замены NаN на ноль мы получили смешанные значения (где-то было True, где-то стоял ноль)


## Расчёты и добавление результатов в таблицу

### Расчет стоимости квадратного метра

```python
df['price_square_meter'] =  df['last_price'] / df['total_area']
```

### Определение даты публикации: день недели, месяц, год.  

```python
df['first_weekday'] = df['first_day_exposition'].dt.weekday # День недели
df['first_month'] = df['first_day_exposition'].dt.month # Месяц
df['first_year'] = df['first_day_exposition'].dt.year # Год
```

### Категоризация объекта по этажам.  

```python
# Функция для проставления категории в зависимости от этажа объекта
def floor_category (row):
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1: 
        return 'Первый'
    if floor == floors_total:
        return 'Последний'
    return 'Другой'
```

```python
df['floor_category'] = df.apply(floor_category, axis=1)
```

### Выведение доли жилой площади и доли кухни в общей площади объекта

Исключим из датасета пустые значения по жилой площади и площади кухни. 

```python
# Удаление пропусков по жилой плоащади
df_notna_area = df.dropna(subset = ['living_area','kitchen_area']).reset_index()
```

```python
df_notna_area.isna().sum()
```

Пропуски в нужных полях удалены. Идем дальше.  

Посчитаем долю, округлим значение до 2 знаков после запятой.

```python
# Доля жилой площади
df_notna_area['share_living_area'] = round(df_notna_area['living_area'] / df_notna_area['total_area'], 2)
```

```python
# Доля площади кухни
df_notna_area['share_kitchen_area'] = round(df_notna_area['kitchen_area'] / df_notna_area['total_area'], 2)
```
